## filtrování

```python
purple_pok = pokemons[pokemons['color'] == 'purple']
purple_pok
```

### pandas logické operátory

    Python         Pandas
    AND            &
    OR             |
    NOT            ~

```python
purple_pok_poison = pokemons[(pokemons['color'] == 'purple') & (pokemons['type 1'] == 'Poison')]
purple_pok_poison
```

```python
green_pok_poison = pokemons[(pokemons['color'] == 'green') & (pokemons['type 1'] == 'Poison')]
green_pok_poison
```

- filtrování na základě seznamu hodnot - .isin

```python
my_colors = ['blue','green','yellow']
```

```python
pok_my_colors =   pokemons[pokemons['color'].isin(my_colors)]
pok_my_colors
```

- filtrování pomocí np.where

```python
strong_poks = pokemons.loc[np.where(pokemons.hp>150)]
strong_poks
```

## spojování df - merge, append

 - spojení dvou df se stejnými sloupci

```python
poison_both_colors = purple_pok_poison.append(green_pok_poison)
poison_both_colors
```

- merge - spojení dvou různých tabulek

```python
weakness = pd.read_excel('data/weakness.xlsx', engine = 'openpyxl')
weakness
```

```python
pok_incl_weakness = pokemons.merge(weakness,how = 'left', left_on = 'type 1', right_on = 'Type')
pok_incl_weakness
```

- úprava sloupců v df, změna pořadí

```python
pok_incl_weakness.columns
```

```python
my_poks = pok_incl_weakness[['name','color','type 1','Weak against']]
my_poks
```

## načtení dat do db - meteodata

```python
with open ('data/meteo_data.csv', encoding='cp1250') as meteo_data:
    obsah = meteo_data.read()
    pozice = obsah.index("Rok;Měsíc;Den;Hodnota;Příznak") + 29
    obsah = obsah[pozice:]

with open ('clean_data.csv', mode = 'w', encoding = "cp1250") as soubor:
        print(obsah, file = soubor)
```

```python

sloupce = ["Rok","Měsíc","Den","Hodnota","Příznak"]
df_from_file = pd.read_csv("clean_data.csv", delimiter = ";", names = sloupce, parse_dates = [[0,1,2]],decimal = ',')
df_from_file['stanice'] = 'stanice01'
df_from_file["datum"] = df_from_file["Rok_Měsíc_Den"].dt.strftime("%Y/%m/%d") 
df_from_file = df_from_file[["stanice","datum","Hodnota","Příznak"]]
df_from_file
```
